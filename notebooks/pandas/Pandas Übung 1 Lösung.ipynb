```python
file_url = "https://opendata.dwd.de/climate_environment/CDC/regional_averages_DE/annual/air_temperature_mean/regional_averages_tm_year.txt"

```

```python
df = pd.read_csv(file_url, sep=';', header=1, index_col='Jahr').drop(['Jahr.1', 'Unnamed: 19'], axis=1)
df.head()
```

Was war der niedrigste und höchste Jahresdurchschitt in Baden-Württemberg? In welchen Jahren sind diese aufgetreten? (Tipp: `max`, `idxmax`)

```python
# Lösung
bw = df['Baden-Wuerttemberg']
print("max BW", bw.max(), bw.idxmax())
print("min BW", bw.min(), bw.idxmin())
```

<hr>Geben Sie den höchsten und niedrigsten Durchschnittswert und das Jahr des Auftretens für alle Spalten aus.

```python
# Lösungen
for colname in df:
    d = df[colname]
    print(colname, '\tMax', d.max(), d.idxmax(), 'Min:', d.min(), d.idxmin())
```

```python
# oder mit apply
def column_to_dict(col):
    return {
        'min': col.min(), 
        'min_year': col.idxmin(),
        'max': col.max(), 
        'max_year': col.idxmax()
    }

df.apply(column_to_dict, result_type='expand').T
```
